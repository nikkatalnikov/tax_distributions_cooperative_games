```python
from functools import reduce
from typing import Tuple, List
import numpy as np
import plotly.graph_objects as go

def order_points(points):
	center = reduce(lambda a, b: (a[0] + b[0], a[1] + b[1]), points, (0, 0))
	center = (center[0] / len(points), (center[1] / len(points)))
	points.sort(key=lambda a: np.math.atan2(a[1] - center[1], a[0] - center[0]))
	return points

def run_task(task_name: str,matrix: List[Tuple[float]], uv: Tuple[float], Nash_point: Tuple[float], threat_point: Tuple[float]):
	print(task_name)
	print("Non-coalition payoff", uv)
	print("Nash's point", Nash_point)
	print("Game with threatens point", threat_point)
	xx, xy, yx, yy = order_points(matrix)
	fig = go.Figure()

	fig.update_xaxes(range=[-2 + np.min([i[0] for i in matrix]), 2 + np.max([i[0] for i in matrix])])
	fig.update_yaxes(range=[-2 + np.min([i[1] for i in matrix]), 2 + np.max([i[1] for i in matrix])])

	fig.update_layout(
		xaxis = dict(
			dtick = 1
		),
		yaxis = dict(
			dtick = 1
		)
	)

	fig.add_shape(
	dict(
		type="line",
		x0=xx[0],
		y0=xx[1],
		x1=xy[0],
		y1=xy[1],
		line=dict(
			color="RoyalBlue",
			width=1
		)
	))

	fig.add_shape(
		type="line",
		x0=xy[0],
		y0=xy[1],
		x1=yx[0],
		y1=yx[1],
		line=dict(
			color="RoyalBlue",
			width=1,
		),
)
	fig.add_shape(
		type="line",
		x0=yx[0],
		y0=yx[1],
		x1=yy[0],
		y1=yy[1],
		line=dict(
			color="RoyalBlue",
			width=1,
		)
	)
	fig.add_shape(
		type="line",
		x0=yy[0],
		y0=yy[1],
		x1=xx[0],
		y1=xx[1],
		line=dict(
			color="RoyalBlue",
			width=1,
		)
	)

	# Create scatter trace of text labels
	fig.add_trace(go.Scatter(
		x=[uv[0], Nash_point[0], threat_point[0]],
		y=[uv[1], Nash_point[1], threat_point[1]],
		text=["Non-coalition payoff", "Nash's point", "threaten point"],
		textposition="top right",
		mode="markers+text",
		marker=dict(
			size=10,
			color=np.random.randn(500), #set color equal to a variable
			colorscale='Reds', # one of plotly colorscales
			showscale=True
		)
	))

	fig.update_shapes(dict(xref='x', yref='y'))
	fig.show("svg")
```

```python
game_matrix = [(3,9),(2,-2),(-1,-3),(12,3)]
uv = (19/7, 21/17)
Nash_point = (8.68, 5.21)
threat_point =  (10, 13/3)
run_task("Task 1", game_matrix, uv, Nash_point, threat_point)
```
