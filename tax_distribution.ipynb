```python
from typing import List
from kpi.taxes import *

def run_task(B: List[float], C: float) -> None:
	print("B:", B)
	print("C:", C)
	print('\n'.join("Poll tax {:0.0f}: {:0.2f}".format(*k) for k in enumerate(poll_tax(B, C))))
	print('\n'.join("Level tax {:0.0f}: {:0.2f}".format(*k) for k in enumerate(level_tax(B, C))))
	print('\n'.join("Proportional tax {:0.0f}: {:0.2f}".format(*k) for k in enumerate(proportional_tax(B, C))))
	print('\n'.join("N-Core tax {:0.0f}: {:0.2f}".format(*k) for k in enumerate(n_core(B, C))))
	print('\n')
```

```python
print("Task 1.1")
run_task([4, 8, 12, 20, 26], 30)
print("Task 1.2")
run_task([4, 8, 12, 20, 26], 50)
print("Task 1.3")
run_task([6000, 8000, 10000], 15000)
```
